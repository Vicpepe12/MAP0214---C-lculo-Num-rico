```python
def SimpsonDouble(f, x_min, x_max, n, *args):
	x = np.linspace(x_min, x_max, n+1)
	integral = f(x_min, *args) + f(x_max, *args)
	for (j,), y in range(x):
		if j % 2 == 0:
			integral += 4*f(y, *args)
		else:
			integral += 2*f(y, *args)
	integral = (x_max - x_min)/(3*n)*integral
	return integral    
	
def SimpsonFloat(f, x_min, x_max, n):	#* \label{lin:simples-ini} *)
	x = np.linspace(x_min, x_max, n+1)
	integral = np.float32(f(x_min)) + np.float32(f(x_max))
	for (j,), y in range(x):
		if j % 2 == 0:
			integral += np.float32(4*f(y))
		else:
			integral += np.float32(2*f(y))
	integral = np.float32((x_max - x_min)/(3*n)*integral)
	return integral
```

```python
import argparse
import numpy as np
import matplotlib.pyplot as plt
from lmfit.models import LinearModel

def main():
	parser = argparse.ArgumentParser(description='EP 3')
	parser.add_argument('exercicio', type=int, choices=[1, 2, 3], help='Escolhe o exercício a ser resolvido')
	parser.add_argument('-L', action='store_true', help='Imprime no formato certo para passar pro LaTeX')
	global args
	args = parser.parse_args()
	
	if args.exercicio == 1:
		exercicio1()
	elif args.exercicio == 2:
		exercicio2()
	elif args.exercicio == 3:
		exercicio3()

def simpsonSimples(f, xmin, xmax, n):	#* \label{lin:simples-ini} *)
	x = np.linspace(xmin, xmax, n+1)
	integ = np.single(f(xmin)) + np.single(f(xmax))
	for (j,), y in np.ndenumerate(x[1:-1]):
		if j % 2 == 0:
			integ += np.single(4*f(y))
		else:
			integ += np.single(2*f(y))
	integ = np.single((xmax - xmin)/(3*n)*integ)
	return integ	#* \label{lin:simples-fim} *)

def simpsonDupla(f, xmin, xmax, n, *args):
	x = np.linspace(xmin, xmax, n+1)
	integ = f(xmin, *args) + f(xmax, *args)
	for (j,), y in np.ndenumerate(x[1:-1]):
		if j % 2 == 0:
			integ += 4*f(y, *args)
		else:
			integ += 2*f(y, *args)
	integ = (xmax - xmin)/(3*n)*integ
	return integ

def exercicio1():
	p = np.array(range(1, 26))
	xmin = 0
	xmax = 1
	
	I = 7
	
	f = lambda x: 8 - 5*x**4
	
	print('Item a)')
	
	integral_simples = np.array([])
	erro_simples = np.array([])
	print('Precisão Simples')
	if args.L:
		print('p', 'N', 'I_num', 'erro', sep=' &', end=' \\\\\n')
	else:
		print('p', 'N', 'I_num', 'erro', sep='\t')
	for pp in p:
		n = 2**pp
		integ = simpsonSimples(f, xmin, xmax, n)
		integral_simples = np.append(integral_simples, integ)
		erro_simples = np.append(erro_simples, np.abs(integ - I))
		if args.L:
			print(pp, n, integ, np.abs(integ - I), sep=' &', end=' \\\\\n')
		else:
			print(pp, n, integ, np.abs(integ - I), sep='\t')
	
	integral_dupla = np.array([])
	erro_dupla = np.array([])
	print('\nPrecisão Dupla')
	if args.L:
		print('p', 'N', 'I_num', 'erro', sep=' &', end=' \\\\\n')
	else:
		print('p', 'N', 'I_num', 'erro', sep='\t')
	for pp in p:
		n = 2**pp
		integ = simpsonDupla(f, xmin, xmax, n)
		integral_dupla = np.append(integral_dupla, integ)
		erro_dupla = np.append(erro_dupla, np.abs(integ - I))
		if args.L:
			print(pp, n, integ, np.abs(integ - I), sep=' &', end=' \\\\\n')
		else:
			print(pp, n, integ, np.abs(integ - I), sep='\t')
	
	print('\nItem b)')
	
	modelo = LinearModel()
	ajusteSimpson = modelo.fit(data=np.log10(erro_dupla[:11]), x=p[:11])
	ajusteRoundoffSimples = modelo.fit(data=np.log10(erro_simples[5:16]), x=p[5:16])
	ajusteRoundoffDupla = modelo.fit(data=np.log10(erro_dupla[12:]), x=p[12:])
	
	fig, ax = plt.subplots()
	
	fig.suptitle('Erros para método de Simpson para precisão simples\ne dupla em função de $\\log_2$ do número de pontos', size='x-large')
	
	ax.set_xlabel(r'$p$', size='large')
	ax.set_ylabel(r'$\log_{10}$erro', size='large')
	ax.plot(p[erro_simples != 0], np.log10(erro_simples[erro_simples != 0]), '+k', markersize=10, label='Precisão simples')
	ax.plot(p, np.log10(erro_dupla), '2r', markersize=10, label='Precisão dupla')
	
	ax.plot(p[:11], ajusteSimpson.best_fit, '-k', linewidth=0.5)
	alphaSimpson = f'{ajusteSimpson.params["slope"].value/np.log10(2):.2f}({ajusteSimpson.params["slope"].stderr/np.log10(2)*100:1.0f})'
	ax.annotate(f'Simpson\n$\\alpha = {alphaSimpson}$', xy=(0.1, -10), fontsize='large')
	print(f'alpha Simpson: {alphaSimpson}')
	
	ax.plot(p[5:16], ajusteRoundoffSimples.best_fit, '-k', linewidth=0.5)
	alphaSimples = f'{ajusteRoundoffSimples.params["slope"].value/np.log10(2):.2f}({ajusteRoundoffSimples.params["slope"].stderr/np.log10(2)*100:1.0f})'
	ax.annotate(f'Roundoff Simples\n$\\alpha = {alphaSimples}$', xy=(8, -4), fontsize='large')
	print(f'alpha Roundoff Simples: {alphaSimples}')
	
	ax.plot(p[12:], ajusteRoundoffDupla.best_fit, '-k', linewidth=0.5)
	alphaDupla = f'{ajusteRoundoffDupla.params["slope"].value/np.log10(2):.2f}({ajusteRoundoffDupla.params["slope"].stderr/np.log10(2)*100:1.0f})'
	ax.annotate(f'Roundoff Dupla\n$\\alpha = {alphaDupla}$', xy=(14, -12), fontsize='large')
	print(f'alpha Roundoff Dupla: {alphaDupla}')
	ax.legend(loc='center right')
	
	plt.savefig('./Exercicio1.svg')
	if not args.L: plt.show()

def trapezios(f, xmin, xmax, n, *args):
	x = np.linspace(xmin, xmax, n+1)
	h = (xmax - xmin)/n
	integ = f(x[0], *args) + f(x[-1], *args)
	integ += 2*f(x[1:-1], *args).sum()
	integ *= h/2
	return integ

def exercicio2():
	l = 1
	g = 9.80665
	
	func = lambda x, theta: 4*np.sqrt(l/g)/np.sqrt(1 - np.sin(theta/2)**2*np.sin(x)**2)
	
	theta = np.linspace(0, np.pi, 20, endpoint=False)
	T = np.array([])
	if args.L:
		print('theta_0', 'T', sep=' &', end=' \\\\\n')
	else:
		print('theta_0', 'T', sep='\t')
	for i, t in enumerate(theta):
		T = np.append(T, trapezios(func, 0, np.pi/2, 2**8, t))
		if args.L:
			print(t, T[i], sep=' &', end=' \\\\\n')
		else:
			print(t, T[i], sep='\t')
	
	theta = np.linspace(0, np.pi, 100, endpoint=False)
	T = np.array([])
	for i, t in enumerate(theta):
		T = np.append(T, trapezios(func, 0, np.pi/2, 2**8, t))
	
	fig, ax = plt.subplots()
	
	fig.suptitle('Período em função do ângulo\npara um pendulo simples', size='x-large')
	
	ax.set_xlabel(r'$\theta$', size='large')
	ax.set_ylabel('$T$', size='large')
	ax.plot(theta, T, '-k')
	
	plt.savefig('Exercicio2.svg')
	if not args.L: plt.show()

def random(Z, a=16807, c=0, m=2147483647):
	Z = np.fmod(a*Z+c, m)
	U = Z/m
	return Z, U

def exercicio3():
	Z = 12557567
	
	func = lambda x: 1 - x**2
	
	print('Item b)')
	
	N = 100
	n = 0
	i = 0
	while i < N:
		Z, U = random(Z)
		x = U
		Z, U = random(Z)
		y = U
		if y < func(x):
			n += 1
		i += 1
	I = n/N
	print(f'I = {I}')
	
	print('\nItem c)')
	
	p = np.array(range(1, 18))
	Nt = 2**p
	
	N = 100
	if args.L:
		print('Nt', 'I_m', 'sigma', 'sigma_m', sep=' &', end=' \\\\\n')
	else:
		print('Nt', 'I_m', 'sigma', 'sigma_m', sep='\t')
	for M in Nt:
		m = 0
		I = np.array([])
		while m < M:
			n = 0
			i = 0
			while i < N:
				Z, U = random(Z)
				x = U
				Z, U = random(Z)
				y = U
				if y < func(x):
					n += 1
				i += 1
			I = np.append(I, n/N)
			m += 1
		if args.L:
			print(M, I.mean(), I.std(ddof=1), I.std(ddof=1)/np.sqrt(M), sep=' &', end=' \\\\\n')
		else:
			print(M, I.mean(), I.std(ddof=1), I.std(ddof=1)/np.sqrt(M), sep='\t')

main()
```
