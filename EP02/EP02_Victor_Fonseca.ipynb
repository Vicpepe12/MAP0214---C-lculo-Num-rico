### Item b)
Para aplicar o método de Eliminação de Gauss para a resolução do sistema linear, faremos o pivoteamento para gerar uma matriz triangular e então realizar a substituição retroativa nesta nova matriz gerada

```python
import numpy as np
```

```python
A = [[0, 5.2, -1.9],
     [11.5, 0, 1.9],
     [1, -1, -1]]

b = [3.2, 15, 0]

def substituicao_retroativa(A, b):
    ## n é a ordem da matriz A
    n = len(A)
    
    ## vetor x com tamanho n
    x = n * [0] 
    
    for i in range(n-1,-1,-1):
        S = 0
        for j in range(i+1, n):
          S = S + A[i][j] * x[j]
        #substituicao para encontrar x 
        x[i] = (b[i] - S)/A[i][i]
    #determinação do vetor residuo r
    r = np.array(b) - np.dot(np.array(A),np.array(x))

    print(f'x = {x}')
    print(f'Erro = {r}')
    return x, r 

def gauss_pivot(A,b):
    for i in range(len(A)):
        p = abs(A[i][i])
        linha_p = i
        print(f"i ={i}")
        #Encontra o maior valor para o pivô
        for j in range(i+1, len(A)):
            if abs(A[j][i]) > p:
                p = abs(A[j][i])
                linha_p = j    
        #permutação de A
        if linha_p != i:
            A[i], A[linha_p] = A[linha_p], A[i]
        #Permutação de B
            b[i], b[linha_p] = b[linha_p], b[i] 
        #Eliminação gaussiana
        for m in range(i+1, len(A)):
            # Define o multiplicador
            mult = A[m][i]/A[i][i]
            # Faz a eliminação
            for n in range(i,len(A)):
                A[m][n] = A[m][n] - mult*A[i][n]
            b[m] = b[m] - mult*b[i]
        print(f"A = {A}") 
        print(f"b = {b}")                
    
    print(f"FINAL: \nA = {A} \nb = {b}") 

    x, r = substituicao_retroativa(A, b)
    return A, b, x, r
A, b, x, r = gauss_pivot(A, b)        
```

### Item c)

```python
A = [[0, 5.2, -1.9],
     [11.5, 0, 1.9],
     [1, -1, -1]]

b = [3.2, 15, 0]

#Permutação das duas primeiras linhas
A[0] , A[1] = A[1] , A[0]
b[0] ,b[1] = b[1] , b[0]

def gauss_jacobi(A, b, epsilon):
     n = len(A)
     #Cria vetores x e x_anterior de tamanho n
     x = n * [0]
     x_anterior = n * [0]
     
     #Aproximação inicial para x e divisão das linhas por A[i][i]
     for i in range(0,n):
          for j in range(0,n):
               if i != j:                   
                    A[i][j] = A[i][j]/A[i][i] #Divisao de cada linha de A e b por A[i][i]
          b[i] = b[i]/A[i][i]
          x_anterior[i] = b[i]

     erro = np.array(x_anterior) #calcula erro inicial
     k = 0
     while np.max(np.abs(erro)) > epsilon:
          for i in range(0, n):
               S = 0
               for j in range(0, n):
                    if i != j:
                         S = S + A[i][j] * x_anterior[j]
               x[i] = b[i] - S
          erro = np.array(x) - np.array(x_anterior) # novo valor do erro
          print(f"I1, I2, I3 (k = {k}) = {x}")
          print(f"Erro = {erro}\n") 
          k +=1
          x_anterior = x[:] #copia o vetor x novo no x_anterior, para a próxima iteração

     print(f"FINAL (depois de {k-1} iterações):")
     print(f"I1, I2, I3 = {x}")
     print(f"Erro = {erro}")     
     return x, erro

x, erro = gauss_jacobi(A,b,10e-4)
```

### Item d)
O método de Gauss-Seidel é muito semelhante ao Jacobi, só que calcula-se o valor de x(k+1) se aproveitando dos valores já calculados até o momento

```python
A = [[0, 5.2, -1.9],
     [11.5, 0, 1.9],
     [1, -1, -1]]

b = [3.2, 15, 0]

#Permutação das duas primeiras linhas
A[0] , A[1] = A[1] , A[0]
b[0] ,b[1] = b[1] , b[0]

print(A)
def gauss_seidel(A, b, epsilon):
     n = len(A)
     #Cria vetores x e x_anterior de tamanho n
     x = n * [0]
     x_anterior = n * [0]
     
     #Aproximação inicial para x e divisão das linhas por A[i][i]
     for i in range(0,n):
          for j in range(0,n):
               if i != j:                   
                    A[i][j] = A[i][j]/A[i][i] #Divisao de cada linha de A e b por A[i][i]
          b[i] = b[i]/A[i][i]
          x_anterior[i] = b[i]

     erro = np.array(x_anterior) #calcula erro inicial
     k = 0
     while np.max(np.abs(erro)) > epsilon:
          for i in range(0, n):
               S = 0
               for j in range(0, n):
                    if i != j:
                         S = S + A[i][j] * x[j] #ATENÇÃO A MUDANÇA: utilizamos o próprio valor de x já calculado para esta iteração
               x[i] = b[i] - S
          erro = np.array(x) - np.array(x_anterior) # novo valor do erro
          print(f" I1, I2, I3 (k = {k}) = {x}")
          print(f"Erro = {erro}") 
          k +=1
          x_anterior = x[:] #copia o vetor x novo no x_anterior, para a próxima iteração

     print(f"FINAL (depois de {k-1} iterações):")
     print(f" I1, I2, I3 = {x}")
     print(f"Erro = {erro}")     
     return x, erro

x, erro = gauss_seidel(A,b,10e-4)
```
